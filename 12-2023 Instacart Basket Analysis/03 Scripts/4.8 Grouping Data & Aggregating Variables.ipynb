## 3. Aggregating Data with transform()

### a) Split the data into groups based on the “user_id” column.
### b) Apply the transform() function on the “quantity” to generate the max orders for user.
### c) Create a new column, “max_order,” into which place the results of the aggregation

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['quantity'].transform(np.max)
```

```python
#checking your aggregation procedure.
pd.options.display.max_rows = None
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check frequency for loyalty_flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## 6. Create a spending flag for each user based on the average price

```python
# Avg of prices for user_id
ords_prods_merge.groupby('user_id')['prices'].mean()
ords_prods_merge['avg_payment'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#checking aggregation procedure.
pd.options.display.max_rows = None
ords_prods_merge.head(20)
```

```python
# spender_type_flag - 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_payment'] < 10, 'spender_type_flag'] = 'Low spender'
```

```python
# spender_type_flag - 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_payment'] >= 10, 'spender_type_flag'] = 'High spender'
```

```python
#Check frequency for spender_type_flag
ords_prods_merge['spender_type_flag'].value_counts(dropna = False)
```

```python
# Check results
ords_prods_merge[['user_id', 'avg_payment', 'spender_type_flag']].head(100)
```

## 7. Determine frequent versus non-frequent customers.

```python
# Median of days_since_prior_order for user_id
ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#checking aggregation procedure.
pd.options.display.max_rows = None
ords_prods_merge.head(20)
```

```python
# Frequency Flag - 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Frequency Flag - 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# Frequency Flag - 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#Check frequency for frequency_flag
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check results
ords_prods_merge[['user_id', 'median_days_since_prior_order', 'frequency_flag']].head(100)
```
